```python
import pandas as pd
import seaborn as sns
import plotly.express as px
from scipy.stats import skew
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from hijri_converter import Hijri, Gregorian
import pyreadstat
```

```python
#import of the data
table1_url = 'https://raw.githubusercontent.com/catherinebah/econ-extension/aef38c7955ab7f04844649b1b6ae4d59a2e6430f/datafile.csv'
table1 = pd.read_csv(table1_url,encoding='latin-1')
display(table1)
```

```python
#extracting the number of the day of the month, but it's still a string. ( Pandas Object type)
table1['dateofmonth']=table1['date'].str[:2]
table1['dateofmonth'].dtypes
```

```python
#Converting in to int
table1['dateofmonth']=table1['dateofmonth'].astype(int)
table1['dateofmonth'].dtypes
```

```python
display(table1)
```

```python
h = Gregorian(1982, 12, 2).to_hijri()
```

```python
#converting to hijri date
for i in table1.index:
    year = table1.loc[i,'year']
    month =table1.loc[i,'month']
    day=table1.loc[i,'dateofmonth']
    
    h=Gregorian(year, month, day).to_hijri()
    table1.loc[i,'hijrifulldate']=h
```

```python
display (table1)
```

```python
display (table1)
```

```python
#extracting the month as a string
table1['hijrifulldate']=table1['hijrifulldate'].astype(str)
```

```python
table1['hijrimonth']=table1['hijrifulldate'].str[5:7]
```

```python
#converting to type int
table1['hijrimonth']=table1['hijrimonth'].astype(int)
```

```python
display(table1)
```

```python
table1["hijrimonthyear"] = 488
hijrimonthyear=488
for i in range(1, len(table1)):
    if (table1.iloc[i,-2]!=(table1.iloc[i-1,-2])):
        hijrimonthyear+=1
    table1.iloc[i,-1]=hijrimonthyear

display(table1)
```
