```python
import pandas as pd
import seaborn as sns
import plotly.express as px
from scipy.stats import skew
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
#import of the data
table1_url = '/content/drive/Shareddrives/Econ paper/ludo/data_file1.csv'
table1 = pd.read_csv(table1_url)
display(table1)
```

```python
table_url = '/content/drive/Shareddrives/Econ paper/ludo/data_drone.xlsx'
table = pd.read_excel(table_url, sheet_name="Drone strikes data")
display(table)
```

```python
clean_table = table.copy()
```

```python
clean_table = clean_table.drop(["Strike link", "Unnamed: 13", "Index"], axis=1)
clean_table = clean_table.loc[(clean_table["Date"]<"2012-01-01")]
display(clean_table)
```

```python
sub = clean_table[["Date", "Minimum total people killed",	"Maximum total people killed",	"Minimum civilians reported killed",	
                   "Maximum civilians reported killed", "Minimum reported injured",	"Maximum reported injured"]]
display(sub)
```

```python
sub["Mean total people killed"] = (sub["Minimum total people killed"] + sub["Maximum total people killed"])/2
sub["Mean total civilians killed"] = (sub["Minimum civilians reported killed"] + sub["Maximum civilians reported killed"])/2
sub["Mean injured"] = (sub["Minimum reported injured"] + sub["Maximum reported injured"])/2
display(sub)
```

```python
final = sub[["Date", "Mean total people killed" ]]
final.rename(columns={'Date': 'date', "Mean total people killed" : "vic_us"}, inplace=True)
display(final)
```

```python
final["date"] = final["date"].dt.strftime("%d%b%Y")
final["date"] = final["date"].str.lower()
```

```python
display(final.iloc[10:41,:])
```

```python
dupli =final[final.duplicated(subset="date")]
dupli_list = dupli["date"].unique()
new_df = []

for date in dupli_list:
  new_row = {"date" : date, "vic_us" : final.loc[final["date"]==date].sum()[1]}
  new_df.append(new_row)

fin = final[~final["date"].isin(dupli_list)]

for row in new_df:
  fin = fin.append(row, ignore_index=True)

final = fin
final["occurrence_us"] = 1.0
final["lnvic_us"] = np.log(final["vic_us"]+1)
final["occurrence_us_1"] = 1.0
```

```python
new_merged = table1.merge(final, on="date", how="left")
new_merged[["vic_us", "occurrence_us", "lnvic_us", "occurrence_us_1"]] = new_merged[["vic_us", "occurrence_us", "lnvic_us", "occurrence_us_1"]].fillna(value=0)
new_merged["occurrence_us_t_y"] = 0.0
new_merged["lnvic_us_t_y"] = 0.0
display(new_merged.iloc[:,-6])

for i in range(1,len(new_merged)):
  if (new_merged.iloc[i,-5]==1.0) or (new_merged.iloc[i-1,-5]==1.0):
    new_merged.iloc[i, -2] = 1.0
  new_merged.iloc[i,-1] = np.log(new_merged.iloc[i,-6]+new_merged.iloc[i-1,-6]+1)
display(new_merged)
```
